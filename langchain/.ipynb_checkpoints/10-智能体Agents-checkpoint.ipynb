```python
import os
import json

# Agent imports
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.agents import AgentType

# Tool imports
from langchain.agents import Tool
from langchain.utilities import GoogleSearchAPIWrapper
from langchain.utilities import DuckDuckGoSearchAPIWrapper
from langchain.utilities import TextRequestsWrapper
from langchain import LLMMathChain
```

```python
from dotenv import find_dotenv, load_dotenv
import os
_ = load_dotenv(find_dotenv())
```

```python
# ç§‘å¤§è®¯é£ž LLM
spark_appid = os.environ['spark_appid']
spark_api_secret = os.environ['spark_api_secret']
spark_api_key = os.environ['spark_api_key']

from langchain_community.chat_models import ChatSparkLLM

llm = ChatSparkLLM(
    spark_app_id=spark_appid, spark_api_key=spark_api_key, spark_api_secret=spark_api_secret
)

llm.invoke('ä½ å¥½')
```

```python
# !pip install numexpr

# åˆå§‹åŒ–ä¸‰ä¸ªtoolï¼šæœç´¢é“¾ã€ç½‘é¡µè¯·æ±‚é“¾ã€è®¡ç®—é“¾
search = DuckDuckGoSearchAPIWrapper()
requests = TextRequestsWrapper()
llm_math_chain = LLMMathChain.from_llm(llm=llm, verbose=True)
```

```python
# å®šä¹‰å·¥å…·åˆ—è¡¨
tools = [
    Tool(
        name='Search',
        func=search.run,
        description="ä½¿ç”¨web search"
    ),
    Tool(
        name='Requests',
        func=requests.get,
        description="è¯·æ±‚URL"
    ),
    Tool(
        name='Calc',
        func=llm_math_chain.run,
        description='è®¡ç®—æ•°å­¦'
    )
]
```

```python
# åˆå§‹åŒ–agent
agent = initialize_agent(tools, 
                         llm, 
                         agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, 
                         verbose=True, 
                         return_intermediate_steps=True, 
                         handle_parsing_errors=True)
```

```python
response = agent({'input': "ä¸­å›½é¦–éƒ½æ˜¯ä»€ä¹ˆ"})
response['output']
```
